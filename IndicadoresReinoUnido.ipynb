```python
##Importar librerias de donde se consulta la informacion
#!pip install world_bank_data
import world_bank_data as wb
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##Definicion de variables
sns.set(style="whitegrid")

# Países de interés
countries = ["GB", "US", "MX", "EU"]
```

```python
## Consulta de informacion de poblacion y grafico
# Indicador de población total
indicator = "SP.POP.TOTL"

# Descargar datos de los últimos 25 años
data = wb.get_series(indicator, country=countries, date="2000:2025")

# Reorganizar en un DataFrame
df = data.unstack(level=0)  # columnas = países, filas = años
df.index = df.index.droplevel("Series")

# Graficar
df_millions = df / 1e6

df_millions.plot(figsize=(10,6))

plt.title("Población (2000-2024)")
plt.ylabel("Millones de habitantes")  # 🔧 cambio aquí
plt.xlabel("Año")
plt.legend(title="País")
plt.grid(True)

plt.show()
```
